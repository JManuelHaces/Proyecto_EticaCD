### <font color= #2E9AFE>1. Introducción </font>

La desempleo es un factor muy importante para el desarrollo de la economia de un pais.Este da por un desajuste en el mercado laboral, donde la oferta de trabajo (por parte de los trabajadores) es superior a la demanda de trabajo (por parte de las empresas). Por lo tanto decidimos analizar la taza de desempelo que se va tratado en USA.

### <font color= #2E9AFE>2. Descripción de la base de datos </font>

Este conjunto de datos contiene datos de series temporales de la tasa de desempleo en EE. UU. desde enero de 2010 hasta el presente. Contiene registros de la tasa de desempleo basados en la calificación educativa, la raza y el género de los adultos.

Este conjunto de datos también contiene datos de la tasa de desempleo por estado para el año 2020.

El conjunto de datos está limpio y no requiere mucho procesamiento previo. El periodo de tiempo de los datos es del 2000 al 2020.

Dentro de los datos encontramos variables protegidas como Edad y Género.

La base de datos fue recuperada de kaggle: https://www.kaggle.com/datasets/aniruddhasshirahatti/us-unemployment-dataset-2010-2020?select=unemployment_data_us.csv


Variables que contiene el dataset:

 - Associates_Degree

 - White

Nota: Las variables guardan datos de tiempo y númericos.

```python
# Librerias
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargando los datos
df = pd.read_excel("./Data/Procesado/Desempleo_USA_General.xlsx")

# Mostrando los datos
df.head()
```

```python
# Columnas que tenemos en los datos
df.columns
```

```python
# Tipos de datos que tenemos
df.dtypes
```

```python
# Ciclo para hacer Datasets por años
for year in range(2010, 2020):
    # Df filtrado por año
    df_filtrados = df[df.Year == year]
    # Guarda los resultados filtrados como un archivo CSV con el nombre correspondiente
    df_filtrados.to_csv(f"./Data/Procesado/Datasets_Por_Año/Des_{year}.csv", index=False)
    # Imprime información del proceso
    print(f"DataFrame para el año {year}, creado.")
```

### <font color= 'navy'> Gráficas </font>

```python
# Creando la figura y los ejes
fig, ax = plt.subplots()
# Graficando las series de tiempo
ax.plot(df['Fecha'], df['Blanco_Cantidad'], label='Blanco')
ax.plot(df['Fecha'], df['Negro_Cantidad'], label='Negro')
ax.plot(df['Fecha'], df['Asiatico_Cantidad'], label='Asiatico')
ax.plot(df['Fecha'], df['Hispano_Cantidad'], label='Hispano')
# Personalizar el gráfico
ax.set_xlabel('Fecha')
ax.set_ylabel('Valor')
ax.set_title('Series de tiempo por cantidad de población racial')
ax.legend()
# Mostrar el gráfico
plt.show()
```

```python
df0 = pd.read_csv("./Data/Procesado/Datasets_Por_Año/Des_2010.csv")
df5 = pd.read_csv("./Data/Procesado/Datasets_Por_Año/Des_2015.csv")
df9 = pd.read_csv("./Data/Procesado/Datasets_Por_Año/Des_2019.csv")

mean = df0[['Hombre_Cantidad','Mujer_Cantidad']].mean()
mean2 = df5[['Hombre_Cantidad','Mujer_Cantidad']].mean()
mean3 = df9[['Hombre_Cantidad','Mujer_Cantidad']].mean()

#Definir los datos
y1 = [mean[0],mean2[0],mean3[0]]
y2 = [mean[1],mean2[1],mean3[1]]
x2 = [2010,2015,2019]

x = np.arange(len(x2))  # lugar de la etiqueta
width = 0.35  # ancho de las barras


fig, ax = plt.subplots()
rects1 = ax.bar(x - width / 2, y1, width, label='Población promedio Hombre',color='seagreen')
rects2 = ax.bar(x + width / 2, y2, width, label='Población promedio Mujer',color='lightgreen')

# Darle nombre a los ejes
ax.set_ylabel('Pesos')
ax.set_xlabel('Años')
ax.set_title('Promedio poblacional 2010, 2015 y 2019 por género ')
ax.set_xticks(x)
ax.set_xticklabels(x2)
ax.legend()
```
